# 1. Write a python program to create a base class "Shape" with methods to calculate area and perimeter. Then, create derived classes "Circle" and "Rectangle" that inherit from the base class and calculate their respective areas and perimeters. Demonstrate their usage in a program.

```python
import math

class Shape:
    def calculate_area(self):
        pass
    
    def calculate_perimeter(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return math.pi * self.radius ** 2

    def calculate_perimeter(self):
        return 2 * math.pi * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)

# Usage example
if __name__ == "__main__":
    # Create a circle and calculate its area and perimeter
    circle = Circle(5)
    circle_area = circle.calculate_area()
    circle_perimeter = circle.calculate_perimeter()
    print(f"Circle - Area: {circle_area}, Perimeter: {circle_perimeter}")

    # Create a rectangle and calculate its area and perimeter
    rectangle = Rectangle(4, 6)
    rectangle_area = rectangle.calculate_area()
    rectangle_perimeter = rectangle.calculate_perimeter()
    print(f"Rectangle - Area: {rectangle_area}, Perimeter: {rectangle_perimeter}")
```

```python
# Define Question and Quiz classes, initialize quiz questions, create quiz objects, and implement user interaction, scoring, and results display.

class Question:
    def __init__(self, text, options, correct_answer):
        self.text = text
        self.options = options
        self.correct_answer = correct_answer

class Quiz:
    def __init__(self, name, questions):
        self.name = name
        self.questions = questions

    def calculate_score(self, user_answers):
        score = 0
        for question, user_answer in zip(self.questions, user_answers):
            if user_answer == question.correct_answer:
                score += 1
        return score

# Create quiz questions
question1 = Question("What is the capital of France?", ["London", "Berlin", "Paris", "Madrid"], "Paris")
question2 = Question("What is 2 + 2?", ["3", "4", "5", "6"], "4")
question3 = Question("Which planet is known as the Red Planet?", ["Earth", "Mars", "Venus", "Jupiter"], "Mars")

# Create a quiz
quiz1 = Quiz("General Knowledge", [question1, question2, question3])

# User interaction and scoring code would go here

# Display results
user_answers = ["Paris", "4", "Mars"]
score = quiz1.calculate_score(user_answers)
print(f"You scored {score} out of {len(quiz1.questions)} on the {quiz1.name} quiz.")
```

```python
class Person:
    def __init__(self, name, age):
        self.__name = name  # Private attribute for name
        self.__age = age    # Private attribute for age

    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age

    def is_eligible_to_vote(self):
        return self.__age >= 18

# Create a Person object
person = Person("Alice", 25)

# Access the name attribute (private attribute) using the getter method
print(f"Name: {person.get_name()}")

# Access the age attribute (private attribute) using the getter method
print(f"Age: {person.get_age()}")

# Check eligibility for voting
if person.is_eligible_to_vote():
    print("This person is eligible to vote.")
else:
    print("This person is not eligible to vote.")
```

```python
import datetime

# Base class for books
class Book:
    def __init__(self, title, author, publication_year):
        self.__title = title
        self.__author = author
        self.__publication_year = publication_year
        self.__checked_out = False
        self.__due_date = None

    # Getter methods
    def get_title(self):
        return self.__title

    def get_author(self):
        return self.__author

    def get_publication_year(self):
        return self.__publication_year

    def is_checked_out(self):
        return self.__checked_out

    def get_due_date(self):
        return self.__due_date

    # Setter methods
    def check_out(self, patron):
        if not self.__checked_out:
            self.__checked_out = True
            today = datetime.date.today()
            self.__due_date = today + datetime.timedelta(days=14)  # Due in 14 days
            return f"{self.__title} has been checked out to {patron.get_name()}. Due on {self.__due_date}."
        else:
            return f"{self.__title} is already checked out."

    def return_book(self):
        if self.__checked_out:
            self.__checked_out = False
            return f"{self.__title} has been returned."
        else:
            return f"{self.__title} is not checked out."

    def __str__(self):
        return f"{self.__title} by {self.__author} ({self.__publication_year})"


# Subclass for fiction books
class FictionBook(Book):
    def __init__(self, title, author, publication_year, genre):
        super().__init__(title, author, publication_year)
        self.__genre = genre

    def get_genre(self):
        return self.__genre

    def __str__(self):
        return super().__str__() + f" - Fiction ({self.__genre})"


# Subclass for non-fiction books
class NonFictionBook(Book):
    def __init__(self, title, author, publication_year, topic):
        super().__init__(title, author, publication_year)
        self.__topic = topic

    def get_topic(self):
        return self.__topic

    def __str__(self):
        return super().__str__() + f" - Non-Fiction ({self.__topic})"


# Patron class
class Patron:
    def __init__(self, name):
        self.__name = name

    def get_name(self):
        return self.__name

    def __str__(self):
        return self.__name


# Transaction class
class Transaction:
    def __init__(self, book, patron, transaction_type):
        self.__book = book
        self.__patron = patron
        self.__transaction_type = transaction_type
        self.__timestamp = datetime.datetime.now()

    def __str__(self):
        return f"{self.__patron.get_name()} {self.__transaction_type} {self.__book.get_title()} on {self.__timestamp}"


# Create some books
book1 = FictionBook("The Hobbit", "J.R.R. Tolkien", 1937, "Fantasy")
book2 = NonFictionBook("Sapiens", "Yuval Noah Harari", 2011, "History")

# Create a patron
patron1 = Patron("Alice")

# Simulate book transactions
transaction1 = Transaction(book1, patron1, "checked out")
print(transaction1)
print(book1.check_out(patron1))
print(book1.return_book())
transaction2 = Transaction(book1, patron1, "returned")
print(transaction2)

# Simulate book reservations
book3 = FictionBook("To Kill a Mockingbird", "Harper Lee", 1960, "Classic")
transaction3 = Transaction(book3, patron1, "reserved")
print(transaction3)

# Display book information
print(book1)
print(book2)
print(book3)
```

```python
from abc import ABC, abstractmethod

class Courses(ABC):
    @abstractmethod
    def courses(self):
        pass

class Python(Courses):
    def courses(self):
        print("Course is Python")
        
class ML(Courses):
    def courses(self):
        print("Course is ML")

class NLP(Courses):
    def courses(self):
        print("Course is NLP")

# Create objects and call methods
python_course = Python()
ml_course = ML()
nlp_course = NLP()

python_course.courses()  # Output: "Course is Python"
ml_course.courses()      # Output: "Course is ML"
nlp_course.courses()     # Output: "Course is NLP"
```
